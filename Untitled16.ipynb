```python
%%writefile pedirnotas.py
def pedir_nombre():
    return input("Ingrese el nombre del estudiante: ")

def pedir_():
    n1 = float(input("Nota 1: "))
    n2 = float(input("Nota 2: "))
    n3 = float(input("Nota 3: "))
    n4 = float(input("Nota 4: "))
    return n1, n2, n3, n4
```

```python
%%writefile promedio.py
import statistics
def calcular_promedio(n1, n2, n3, n4):
    return (n1 + n2 + n3 + n4) / 4
def mediana(n1, n2, n3, n4):
  return statistics.median([n1, n2, n3, n4])
def moda(n1, n2, n3, n4):
  return statistics.mode([n1, n2, n3, n4])
```

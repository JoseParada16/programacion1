```python
conjunto_A = {1, 2, 3, 3, 5}
conjunto_B = {1, 7, 2, 4, 5}
conjunto_A.difference(conjunto_B)
```

El metodo A.symmetric_difference(B) lo que hace es ver que datos unicos tienen cada uno de los conjuntos (datos que tenga A que no tenga B, y datos que tenga B que no tenga A), y la salida nos da un solo conjunto de estos datos en orden numerico

Código 2: Cree una función cuya entrada sea un objeto tipo str, el cual cuente:
 Vocales mayúsculas y minúsculas,
 Consonantes mayúsculas y minúsculas,
 números y
 otros caracteres (como espacios, números o sı́mbolos)
La salida de la función es un diccionario con la cantidad total de vocales, consonantes y otros
caracteres encontrados en el texto.
Ejecute la función con una cadena ingresada por el usuario.

```python
def contador_vocales1(palabra):
  vocales = "aeiou"
  consonantes = "bcdfghjklmnñpqrstvwxyz"
  contador = 0
  contador1 = 0
  contador2 = 0

  for letra in palabra.lower():
    if letra in vocales:
      contador += 1
    elif letra in consonantes:
      contador1 += 1
    else:
      contador2 += 1
  return({"Cantidad de vocales:": contador,
          "Cantidad de consonates:": contador1,
          "Cantidad de otros caracteres:": contador2})
```
